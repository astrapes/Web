```python
total_review = int(soup.find('strong','total').find('em').string.replace(',',''))
total_review
```

```python
loop_total_review = math.ceil(total_review/10)
loop_total_review
```

```python
score = []
date = []
reple = []

for i in range(1,loop_total_review+1):
    url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=186615&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
    html = urlopen(url + str(i))
    soup = BeautifulSoup(html,'html.parser')
    lists = soup.find_all(class_ = 'score_result')
    lis = lists[0].find_all('li')
    for li in lis:
        score.append(int(li.find(class_='star_score').find('em').get_text()))
        date.append(li.find_all('em')[2].get_text())
        count_span = len(li.find(class_='score_reple').find_all('span'))
        viewer = li.find(class_ = 'ico_viewer')
        spoiler = li.find(class_ = 'text_spo _text_spo')
        if viewer != None or spoiler != None:
            reple.append(li.find(class_ = 'score_reple').find('p').get_text().replace('스포일러가 포함된 감상평입니다. 감상평 보기','').strip('\n\t\r관람객'))
        elif count_span == 3:
            reple.append(li.find('a')['data-src'])
        else :
            reple.append(li.find(class_ = 'score_reple').find('p').get_text().strip())
```

```python
df = pd.DataFrame(dict(zip(['평점','날짜','리플'],[score,date,reple])))
```
